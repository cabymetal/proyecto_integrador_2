```python
import datetime
columnas_valores_numericos = ['CantidadPQRTotal', 'PromedioConsumoAguas','ValorFacturadoAguasXConsumo','ValorFacturadoAlcantarilladoXConsumo','ValorFacturadoAseo',
                              'ValorFacturadoEnergiaXConsumo','ValorFacturadoGasXConsumo','ValorServiciosFinancieros',
                              'DesviacionConsumoMedioAguas','DesviacionConsumoMedioEnergia','DesviacionConsumoMedioGas','ValorCompensadoEnergia','FrecuenciaCompensacionEnergia',
                              'ValorTotalIngresosUsuario', 'MaximoNumeroCuentasVencidasVigentesEnergia', 'MaximoNumeroCuentasVencidasVigentesGas',
                              'NumeroCuentasVencidasVigentesAguas','NumeroCuentasVencidasVigentesEnergia', 'NumeroCuentasVencidasVigentesGas',
                              'BalanceTransformadorMaximo', 'BalanceTransformadorPorcentaje' , 'BalanceTransformadorMinimo']
fill_dict = {x: 0 for x in columnas_valores_numericos}

df['TieneServicioAgua'] = ((df['NumeroPeriodosConsumoAguas'].isna()) | (df['PromedioConsumoAguasUltimos24Meses'].isna() | df['NumeroPeriodosConsumoAguas'].isna()))
df['TieneServicioAlcantarillado'] = ((df['PromedioVertimientoAlcantarillado'].isna()) | (df['PromedioVertimientoAlcantarilladoUltimos24Meses'].isna() | df['NumeroPeriodosConsumoAlcantarillado'].isna()))
df['TieneServicioEnergia'] = ((df['PromedioConsumoEnergia'].isna()) | (df['PromedioConsumoEnergiaUltimos24Meses'].isna() | df['NumeroPeriodosConsumoEnergia'].isna()))
df['TieneServicioGas'] = ((df['PromedioConsumoGas'].isna()) | (df['PromedioConsumoGasUltimos24Meses'].isna() | df['NumeroPeriodosConsumoGas'].isna()))
df['HaEnviadoPQR'] = df.loc[:, 'CantidadPQREscrito':'CantidadPQRWeb'].sum(axis=1) > 0
df['ConsumeMasMedianaMunicipioAgua'] = (df['PromedioConsumoAguas']  >= df['MedianaMunicipioEstratoAgua'])
df['ConsumeMasMedianaMunicipioEnegia'] = (df['PromedioConsumoEnergia']) >= df['MedianaMunicipioEstratoEnergia']
df['ConsumeMasMedianaMunicipioGas'] = (df['PromedioConsumoGas'] >= df['MedianaMunicipioEstratoGas'])
df2 = df.loc[:, 'CantidadCortesAguas': 'CantidadSuspensionesGas'].fillna(0)
df.loc[:, 'CantidadCortesAguas': 'CantidadSuspensionesGas'] = df2
#df = df[df['BalanceTransformadorMaximo'] >= 0] #valores negativos pueden estar mal calculados validar
df = df[df['BalanceTransformadorPorcentaje'] >= 0] #porcentajes menores a cero no son validos
#df = df[df['BalanceTransformadorMinimo'] >= 0]
df2 = df.loc[:, columnas_valores_numericos].fillna(fill_dict)
df.loc[:, columnas_valores_numericos] = df2
df = df.loc[~df['TotalPropPersonasNBI'].isna()]
df = df.loc[~df['TotalComponenteDependenciaEconomicaNBI'].isna()] #quitar filas con valores no válidos
df = df.loc[~df['MaxScoreCartera'].isna()] #quitar filas con valores nulos
df['FechaAntiguedadInstalacion'] = pd.to_datetime(df['FechaAntiguedadInstalacion'])
df['Antiguedad'] = datetime.datetime.now() - df['FechaAntiguedadInstalacion']


del df2
```

```python
import numpy as np
from scipy import stats

colums_to_check = ['Altitud', 'TotalPropPersonasNBI',
       'TotalComponenteDependenciaEconomicaNBI', 'CantidadVisitasReconexion',
       'CantidadVisitasSuspension', 'ValorFacturadoAguasXConsumo', 'ValorFacturadoAlcantarilladoXConsumo',
       'ValorFacturadoAseo', 'ValorFacturadoEnergiaXConsumo',
       'ValorFacturadoGasXConsumo', 'ValorServiciosFinancieros','ValorCompensadoEnergia', 'FrecuenciaCompensacionEnergia',
       'ValorTotalIngresosUsuario', 'CantidadPQRTotal', 'MaximoNumeroCuentasVencidasVigentesEnergia',
       'MaximoNumeroCuentasVencidasVigentesGas',
       'MaximoNumeroCuentasVencidasVigentesInstalacion',
       'NumeroCuentasVencidas']

z_scores= np.abs(stats.zscore(df.loc[:, colums_to_check]))

df_final = df[(z_scores < 3).all(axis=1)]
df.loc[(z_scores > 3).all(axis=1), colums_to_check].head() #muestra filas que en alguna columna tienen un zscore mayor a 3
#es decir 3 desviaciones estandar por encima de la media son outliers
#en este caso no hubieron concluimos que esta base es la que se va a utilizar en la segmentación
```

```python
df_final.shape
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=91,svd_solver= 'randomized')
pca.fit(df_final.select_dtypes(['number', 'bool'], exclude= 'timedelta'))
pca.explained_variance_ratio_
```

```python
pca.singular_values_
```

podemos seleccionar ya un porcentaje de variables que queramos utilizar por ejemplo si seleccionamos las variables que expliquen el 99.9999999% de los datos hacemos lo siguiente

```python
acum_variance = 0 
idx = 0
for idx, val in enumerate(pca.explained_variance_ratio_):
    acum_variance = acum_variance + val
    if acum_variance >= 0.999999999999:
        idx = idx
        break

idx
```

nos indica que tendríamos que seleccionar hasta la columna 78


```python
df_export = df_final.select_dtypes(['number', 'bool'], exclude= 'timedelta').iloc[:, range(0,78)]
df_export = pd.concat([df_export,df_final.select_dtypes(['timedelta'])], axis=1)
df_export['Antiguedad'] = df_export['Antiguedad'].dt.days
df_export.shape
```
